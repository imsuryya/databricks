```python
employee_csv   = "/Volumes/assingment/default/employee/Employee-Q1(in).csv"
department_csv = "/Volumes/assingment/default/department/Department-Q1(in).csv"
country_csv    = "/Volumes/assingment/default/country/Country-Q1(in).csv"
```

```python
employee_df   = read_csv(employee_csv)
department_df = read_csv(department_csv)
country_df    = read_csv(country_csv)
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

employee_schema = StructType([
    StructField("EmployeeID",   IntegerType(), True),
    StructField("EmployeeName", StringType(),  True),
    StructField("Department",   StringType(),  True),
    StructField("Country",      StringType(),  True),
    StructField("Salary",       IntegerType(), True),
    StructField("Age",          IntegerType(), True)
])

department_schema = StructType([
    StructField("DepartmentID",   StringType(), True),
    StructField("DepartmentName", StringType(), True)
])

country_schema = StructType([
    StructField("CountryCode", StringType(), True),
    StructField("CountryName", StringType(), True)
])
```

```python
employee_df = (spark.read
                    .format("csv")
                    .schema(employee_schema)
                    .option("header", "true")
                    .load(employee_csv))

department_df = (spark.read
                      .format("csv")
                      .schema(department_schema)
                      .option("header", "true")
                      .load(department_csv))

country_df = (spark.read
                   .format("csv")
                   .schema(country_schema)
                   .option("header", "true")
                   .load(country_csv))
```

```python
employee_df.display()
department_df.display()
country_df.display()
```

```python
write_csv(employee_df,   "/Volumes/assingment/bronze/emp/")
write_csv(department_df, "/Volumes/assingment/bronze/department/")
write_csv(country_df,    "/Volumes/assingment/bronze/country/")
```
