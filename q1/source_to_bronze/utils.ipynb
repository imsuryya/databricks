```python

def read_csv(path):
    return (spark.read
                 .format("csv")
                 .option("header", "true")
                 .load(path))
```

```python
def write_csv(df, path, mode="overwrite"):
    (df.write
       .format("csv")
       .mode(mode)
       .option("header", "true")
       .save(path))
```

```python
def write_delta(df, path, mode="overwrite"):
    (df.write
       .format("delta")
       .mode(mode)
       .save(path))
```

```python
def read_delta(path):
    return spark.read.format("delta").load(path)
```
